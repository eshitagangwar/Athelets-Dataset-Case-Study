```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ath=pd.read_csv("athlete_events.csv")
```

```python
ath100=ath.head(100)
ath100
```

### which season is most popular?

```python
sea=ath.groupby(['Season']).size()
sea=pd.DataFrame(sea)
sea=sea.reset_index()
sea=sea.rename(columns={0:'Count'})
sea
```

```python
plt.pie(sea['Count'],labels=sea['Season'],autopct='%1.2f%%',startangle=90)
plt.axis("equal")
plt.show()
```

### which are the cities which have hosted olympics?which city has hosted the maximum number of olympics?

```python
ath100
```

```python
host=ath.groupby(['City','Year']).size()
host=pd.DataFrame(host)
host=host.reset_index()
host=host.rename(columns={0:'Count'})
host
```

```python
host1=host.groupby(['City']).size()
host1=pd.DataFrame(host1)
host1=host1.reset_index()
host1=host1.rename(columns={0:'Count'})
host1
```

```python
fig=plt.figure(figsize=(12,7))
plt.bar(host1['City'],host1['Count'],color='red')
plt.xticks(rotation='vertical')
plt.xlabel("city")
plt.ylabel("count")
plt.title("hosted olympics")
plt.show()
```

## top 10 teams having maximum number of athletes?

```python
maxath=ath.groupby(['Team','Name']).size()
maxath=pd.DataFrame(maxath)
maxath=maxath.reset_index()
maxath=maxath.rename(columns={0:'Count'})
maxath
```

```python
max1=maxath.groupby(['Team']).size()
max1=max1.sort_values(ascending=False)
max1=pd.DataFrame(max1)
max1=max1.reset_index()
max1=max1.rename(columns={0:'Count'})
max1=max1.head(10)
max1
```

```python
fig=plt.figure(figsize=(12,7))
plt.bar(max1['Team'],max1['Count'])
plt.xticks(rotation='vertical')
plt.xlabel("Team")
plt.ylabel("Count")
plt.title
```

### top 10 teams having maximum average height of athletes?

```python
ath1=ath.dropna(subset=['Height'])
```

```python
maxht=ath1.groupby(['Team','Name'])['Height'].mean()
maxht=maxht.sort_values(ascending=False)
maxht=pd.DataFrame(maxht)
maxht=maxht.reset_index()
maxht
```

```python
maxht1=maxht.groupby(['Team'])['Height'].mean()
maxht1=maxht1.sort_values(ascending=False)
maxht1=pd.DataFrame(maxht1)
maxht1=maxht1.reset_index()
maxht1=maxht1.head(10)
maxht1
```

```python
fig=plt.figure(figsize=(12,7))
plt.bar(maxht1['Team'],maxht1['Height'])
plt.xticks(rotation='vertical')
plt.xlabel("Team")
plt.ylabel("avg height")
plt.title("top 10 teams of max avg height")
plt.show()
```

### top 10 countries having received maximum gold
